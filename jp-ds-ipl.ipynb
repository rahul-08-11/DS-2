### reading dataset
source:https://www.mykhel.com/cricket/ipl-2022-stats-s4/#T20#tab-mostrun


```python
r = requests.get(url, headers=header)
df = pd.read_html(r.text)
df=df[0]
del df["Unnamed: 1"]
df.set_index(keys="POS",inplace=True)
df
```

## All teams performance in single space

```python
x=np.array(df["INN"]).reshape(27,6)
```

```python
y=np.array(df["RUNS"]).reshape(27,6)
y1=np.array(df["SR"]).reshape(27,6)
```

```python

team_color={"Rajasthan":"green",
           "Lucknow":"yellow",
           "Delhi":"blue",
           "Gujarat":"red",
           "Hyderabad":"cyan",
           "Mumbai":"violet",
           "Kolkata":"aquamarine",
           "Punjab":"lime",
           "Bangalore":"magenta",
           "Chennai":"indigo"}
fig,ax=plt.subplots(2,1,figsize=(30,30))
ax[0].tick_params(axis='x', labelsize=25)
ax[0].tick_params(axis='y', labelsize=25)
ax[1].tick_params(axis='x', labelsize=25)
ax[1].tick_params(axis='y', labelsize=25)
for i in range(1,len(x)):
    ax[0].scatter(y[i],x[i],color=team_color[df["TEAM"][i]])
    ax[1].scatter(y1[i],x[i],color=team_color[df["TEAM"][i]])

```

### From the above analysis,we could say that rajasthan,gujarat and lucknow team has performed very well as they are seen to be holding the top areas in the plot.

### Overall best  bowling  as well as batting performance of bowlers

```python
df_bowling["Runs"]=df_bowling["Runs"].astype(int)
players_bbps=[]
for i in range(48):
    bbp=df_bowling["Runs"][i]*df_bowling["Wkts"][i]/100
    players_bbps.append(bbp)
 
```

```python

df_new=df_bowling[["Player","Mat","Runs","Wkts"]]
df_new.insert(3,"bbps",players_bbps)
```

```python
df_new_top=df_new.tail(10)
df_new_top
```
